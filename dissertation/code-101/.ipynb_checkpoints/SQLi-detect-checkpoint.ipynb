```python
def create_sample_data(n_samples=10000):
    """
    Create sample SQL injection data for demonstration purposes.
    This simulates the structure of the RbSQLi dataset.
    
    Reference: Data structure should match image.png columns:
    - sql_query: The SQL query string
    - injection_type: Type of injection (None_Type, Error-Based, meta_based, Time-Based)
    - vulnerability_status: Yes/No for whether it's an injection attack
    - sql_command, target_table, selected_columns, etc.
    """
    np.random.seed(SEED)
    
    # Define benign SQL queries
    benign_queries = [
        "SELECT * FROM users WHERE id = 1",
        "SELECT name, email FROM customers WHERE age > 18",
        "INSERT INTO orders (product_id, quantity) VALUES (101, 5)",
        "UPDATE products SET price = 29.99 WHERE product_id = 100",
        "DELETE FROM sessions WHERE created_at < '2024-01-01'",
        "SELECT COUNT(*) FROM transactions WHERE status = 'completed'",
        "SELECT user_id, order_id FROM payments WHERE amount BETWEEN 10 AND 100",
        "CREATE TABLE temp_data (id INT, name VARCHAR(100))",
        "ALTER TABLE users ADD COLUMN phone_number VARCHAR(15)",
        "DROP INDEX idx_email ON customers"
    ]
    
    # Define malicious SQL injection patterns
    malicious_queries = {
        'Union-based': [
            "SELECT * FROM users WHERE id = 1 UNION SELECT username, password FROM admin",
            "SELECT name FROM products WHERE id = 1 UNION ALL SELECT credit_card FROM payments",
            "SELECT email FROM customers WHERE id = 1 UNION SELECT * FROM secret_data",
            "SELECT * FROM login WHERE user = 'admin' UNION SELECT 1,2,3,4,5,6,7,8,9,0",
        ],
        'Error-based': [
            "SELECT * FROM users WHERE id = 1 AND 1=CONVERT(int, @@version)",
            "SELECT * FROM products WHERE name = 'test' AND 1=CONVERT(int, (SELECT @@version))",
            "SELECT * FROM admin WHERE id = 1 AND (SELECT COUNT(*) FROM users) > 0",
            "SELECT * FROM employees WHERE name LIKE '%' AND 1=CAST('a' AS INT)--"
        ],
        'Boolean-based': [
            "SELECT * FROM users WHERE id = 1 AND 1=1",
            "SELECT * FROM admin WHERE username = 'admin' AND '1'='1'",
            "SELECT * FROM payments WHERE amount > 100 AND 1=1 OR 1=1",
            "SELECT * FROM login WHERE user_id = 1 AND (SELECT 1 FROM dual WHERE 1=1) IS NOT NULL"
        ],
        'Time-based': [
            "SELECT * FROM users WHERE id = 1 AND SLEEP(5)",
            "SELECT * FROM admin WHERE username = 'admin' AND BENCHMARK(10000000, MD5('a'))",
            "SELECT * FROM payments WHERE order_id = 1 AND pg_sleep(10)",
            "SELECT * FROM employees WHERE name = 'test' AND (SELECT COUNT(*) FROM users) > 0 AND SLEEP(3)"
        ],
        'Stack-queries-based': [
            "SELECT * FROM users WHERE id = 1; DROP TABLE users;",
            "SELECT * FROM admin WHERE user = 'admin'; DELETE FROM logs;",
            "SELECT * FROM payments WHERE amount > 100; INSERT INTO hacker (data) VALUES ('stolen');",
            "SELECT * FROM employees; TRUNCATE TABLE salaries;"
        ],
        'Meta-based': [
            "SELECT * FROM users WHERE id = 1 AND 1=0 UNION SELECT table_name FROM information_schema.tables",
            "SELECT * FROM admin WHERE id = 1 AND 1=1 ORDER BY (SELECT NULL UNION SELECT NULL UNION SELECT NULL)",
            "SELECT * FROM payments WHERE amount = 1 AND (SELECT COUNT(*) FROM (SELECT 1 UNION SELECT 2) AS t) > 0"
        ]
    }
    
    data = []
    
    # Generate benign samples (70% approximately)
    n_benign = int(n_samples * 0.7)
    for i in range(n_benign):
        query = np.random.choice(benign_queries)
        data.append({
            'sql_query': query,
            'injection_type': 'None_Type',
            'vulnerability_status': 'No',
            'sql_command': query.split()[0].upper() if query.split() else 'SELECT',
            'target_table': query.lower().split('from')[-1].split()[0] if 'from' in query.lower() else 'unknown',
            'selected_columns': '*',
            'comparison_operator': '=',
            'logical_operator': 'AND',
            'sql_comment_syntax': ''
        })
    
    # Generate malicious samples (30% approximately)
    injection_types = list(malicious_queries.keys())
    for injection_type in injection_types:
        n_malicious = int(n_samples * 0.3 / len(injection_types))
        for i in range(n_malicious):
            query = np.random.choice(malicious_queries[injection_type])
            data.append({
                'sql_query': query,
                'injection_type': injection_type,
                'vulnerability_status': 'Yes',
                'sql_command': query.split()[0].upper() if query.split() else 'SELECT',
                'target_table': query.lower().split('from')[-1].split()[0] if 'from' in query.lower() else 'unknown',
                'selected_columns': '*',
                'comparison_operator': '=',
                'logical_operator': 'AND',
                'sql_comment_syntax': '--' if '--' in query else ''
            })
    
    df = pd.DataFrame(data)
    print(f"\n✓ Sample data created with {len(df)} rows")
    print(f"  Benign: {n_benign} ({(n_benign/len(df))*100:.1f}%)")
    print(f"  Malicious: {len(df)-n_benign} ({((len(df)-n_benign)/len(df))*100:.1f}%)")
    
    return df
```
