### T2 Use the simple linear regression algorithm on the data set water_data.csv

 - T_degC: Temperature in degrees Celsius
 - Salnty: Salinity

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt # MATLAB-like way of plotting
from mpl_toolkits.mplot3d import Axes3D
# sklearn package for machine learning in python:
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv('water_data.csv') # read data (make sure .csv in folder)
```

```python
print(df.head(),'\n') # print first 5 rows of data
```

```python
#X = df.iloc[:, [0,1]].values # inputs
#y = df.iloc[:, 1].values # target

# Convert to DataFrame
df = pd.DataFrame(df)

# 2. Prepare the features and labels
X = df[['T_degC']]  # must be 2D for sklearn
y = df['Salnty']
```

```python
# split the data into training and test sets:
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/4, random_state = 42)
```

```python
# fit the linear least-squares regression line to the training data:
regr = LinearRegression()
regr.fit(X_train, y_train)
# 5. Make predictions
y_pred = regr.predict(X_test)
```

```python
# 6. Print results
print("Coefficient (slope):", regr.coef_[0])
print("Intercept:", regr.intercept_)
print("Mean Squared Error (MSE):", mean_squared_error(y_test, y_pred))
print("R² Score:", r2_score(y_test, y_pred))
```

```python
# 7. Plot the regression line
plt.figure(figsize=(10,6))
plt.scatter(X, y, color='blue', label='Actual data')
plt.plot(X_test, y_pred, color='red', linewidth=2, label='Regression Line')
plt.xlabel('Temperature (°C)')
plt.ylabel('Salinity')
plt.title('Simple Linear Regression: Temperature vs Salinity')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Interpretation

The regression line shows how salinity (Salnty) changes with temperature (T_degC).

The coefficient (slope) tells you how much Salnty increases or decreases for each °C of temperature.

The intercept is the expected salinity when temperature is 0°C.

The R² score indicates how well the model explains the variation in salinity.
