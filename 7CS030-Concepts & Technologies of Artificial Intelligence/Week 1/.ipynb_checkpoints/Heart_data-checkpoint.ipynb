```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt # MATLAB-like way of plotting
from mpl_toolkits.mplot3d import Axes3D
# sklearn package for machine learning in python:
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv('heart_data.csv') # read data (make sure .csv in folder)
```

```python
print(df.head(),'\n') # print first 5 rows of data
```

```python
X = df.iloc[:, [0,1]].values # inputs
y = df.iloc[:, 2].values # target
```

```python
# split the data into training and test sets:
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/4, random_state = 0)
```

```python
# fit the linear least-squares regression line to the training data:
regr = LinearRegression()
regr.fit(X_train, y_train)
```

```python
# The coefficients
print('Coefficients: ', regr.coef_)

# The coefficients
print('Intercept: ', regr.intercept_)

# The mean squared error
print('Mean squared error: %.8f'
% mean_squared_error(y_test, regr.predict(X_test)))

# The coefficient of determination: 1 is perfect prediction
print('Coefficient of determination: %.2f'
% r2_score(y_test, regr.predict(X_test)))
```

```python
# visualise initial data set in 3D
fig1 = plt.figure(figsize=(8,6))
ax1 = fig1.add_subplot(111, projection = '3d')
ax1.scatter(X[:,0], X[:,1], y, color = 'blue')
ax1.azim = -60
ax1.dist = 10
ax1.elev = 20
ax1.set_xlim(0,80)
ax1.set_ylim(0,30)
ax1.set_zlim(0,20)
ax1.set_title('Initial Dataset')
ax1.set_xlabel('Biking to Work (% population)')
ax1.set_ylabel('Smoking (% population)')
ax1.set_zlabel('Heart Disease (% population)')
```

```python
# visualise training set results in 3D:
fig2 = plt.figure(figsize=(8,6))
ax2 = fig2.add_subplot(111, projection = '3d')

# plot the training data
ax2.scatter(X_train[:,0], X_train[:,1], y_train, color = 'blue')

# plot the plane which represents the model
X1, X2 = np.meshgrid(range(80), range(30))
Z = regr.coef_[0]*X1+regr.coef_[1]*X2+regr.intercept_
ax2.plot_surface(X1, X2, Z, alpha=0.5, color='red')
ax2.azim = -60
ax2.dist = 10
ax2.elev = 20
ax2.set_xlim(0,80)
ax2.set_ylim(0,30)
ax2.set_zlim(0,20)
ax2.set_title('Training set')
ax2.set_xlabel('Biking to Work (% population)')
ax2.set_ylabel('Smoking (% population)')
ax2.set_zlabel('Heart Disease (% population)')
fig2.tight_layout(pad=-2.0)
```

```python
# visualise test set results in 3D with plane:
fig3 = plt.figure(figsize=(8,6))
ax3 = fig3.add_subplot(111, projection = '3d')

# plot the test data
ax3.scatter(X_test[:,0], X_test[:,1], y_test, color = 'blue')

# plot the plane which represents the model
X1, X2 = np.meshgrid(range(80), range(30))
Z = regr.coef_[0]*X1+regr.coef_[1]*X2+regr.intercept_
ax3.plot_surface(X1, X2, Z, alpha=0.5, color='red')
ax3.azim = -60
ax3.dist = 10
ax3.elev = 20
ax3.set_xlim(0,80)
ax3.set_ylim(0,30)
ax3.set_zlim(0,20)
ax3.set_title('Test set')
ax3.set_xlabel('Biking to Work (% population)')
ax3.set_ylabel('Smoking (% population)')
ax3.set_zlabel('Heart Disease (% population)')
fig3.tight_layout(pad=-2.0)
```

### What Does the Plane Represent?

The regression plane is the best fit surface that the model learned. For any given pair of input features
x1 and x2, the model predicts: ^y​=β0​⋅x1​+β1​⋅x2​+intercept
