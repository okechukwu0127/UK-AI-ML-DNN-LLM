#### Goal of the Code

**The program:**

- Loads the wine dataset (chemical measurements + cultivar label)
- Encodes the features into numeric form
- Splits the data into training and testing sets
- Trains a multi-layer neural network (MLPClassifier)
- Evaluates model performance (accuracy + misclassified points)

It's an example of supervised classification using neural networks.

This code is a complete example of using a neural network (MLPClassifier) from scikit-learn to classify wine samples based on their chemical composition.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import LabelEncoder
```

```python
pd.set_option('display.max_columns', None)
df = pd.read_csv('wine_data.csv')
print(df.head())
```

#### 3️⃣ Encode Features Numerically

```python
for i in range(1,14):
    df.iloc[:,i] = LabelEncoder().fit_transform(df.iloc[:,i])
print(df.head(),'\n')

#Loops through columns 1 to 13 (skipping the first, “Cultivator”)
#Applies LabelEncoder to each feature column.
```

#### 4️⃣ Define Inputs (X) and Outputs (y)

```python
X = df.iloc[:,1:14].values
y = df.iloc[:,0].values

#X: all input features (Alcohol → Proline)
#y: target labels (Cultivator) — the type of wine.
```

#### 5️⃣ Split the Dataset

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, random_state=42, test_size=1/3)
```

#### 6️⃣ Define and Train the Neural Network

```python
mlp = MLPClassifier(hidden_layer_sizes=(10,50,20),
                    activation="logistic",
                    random_state=0, max_iter=2000)
mlp.fit(X_train, y_train)
```

**Key parts:**
    
    - hidden_layer_sizes=(10,50,20)
      - The network has 3 hidden layers with 10, 50, and 20 neurons each.
    
    - activation="logistic"
      - Uses the sigmoid activation function → outputs values between 0 and 1.
      - Suitable for binary or multi-class classification.

    - max_iter=2000
      - The maximum number of training iterations to ensure convergence.

#### 7️⃣ Evaluate Performance

```python
print('Our Accuracy is %.2f' % mlp.score(X_test, y_test))
print('Number of mislabeled points out of a total %d points : %d'
      % (X_test.shape[0], (y_test != mlp.predict(X_test)).sum()))
```
