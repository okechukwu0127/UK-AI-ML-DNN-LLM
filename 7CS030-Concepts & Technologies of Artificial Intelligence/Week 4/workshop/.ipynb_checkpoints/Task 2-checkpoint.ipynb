Construct Neural Networks (NN) to model data generated by the make_classification function in Python.
- a) Data has 100 samples, 1 feature, 2 classes and the NN has no hidden layers.
- Compare the results with using logistic regression. 
- Is there a difference? 
- Add some hidden layers, see if you can improve the performance.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import Normalize, ListedColormap
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_classification, make_moons, make_circles
```

```python

X, y = make_classification(n_samples = 100, n_features = 1, n_redundant = 0, n_informative = 1, 
                           n_classes = 2, n_clusters_per_class = 1, flip_y = 0.3, random_state = 42)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=1/3, random_state=0)
```

```python
#mlp = MLPClassifier(hidden_layer_sizes=(50,100,50),
#                    activation="tanh", random_state=1, max_iter=2000)


mlp = MLPClassifier(hidden_layer_sizes = (), activation = 'logistic', max_iter = 2000,
                   random_state = 42) 

mlp.fit(X_train, y_train)
```

```python
print('Our Accuracy is %.2f' % mlp.score(X_test, y_test))
print('Number of mislabeled points out of a total %d points : %d'
      % (X_test.shape[0], (y_test != mlp.predict(X_test)).sum()))
```

```python
#construct model and fit to the training data
logre = LogisticRegression()
logre.fit(X_train, y_train)
```

```python
# output the accuracy score
print('Our Accuracy is %.2f' % logre.score(X_test, y_test))

# output the number of mislabeled points
print('Number of mislabeled points out of a total %d points : %d'
% (X_test.shape[0], (y_test != logre.predict(X_test)).sum()))
```
