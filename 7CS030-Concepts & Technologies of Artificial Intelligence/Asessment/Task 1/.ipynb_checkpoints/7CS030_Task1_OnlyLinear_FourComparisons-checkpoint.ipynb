This notebook implements **four** Linear Regression experiments:
- **Model A:** Single-feature Linear Regression (visual, explanatory baseline).  
- **Model B:** Two-feature Linear Regression (extends the simple model).  
- **Model C:** Linear Regression using the **top 5 features**.  
- **Model D:** Linear Regression using **all numeric features**.  

We evaluate each with **RMSE**, **MAE**, and **R²** on the same train/test split for fairness.  
We include at most **two figures**:
1) Figure 1 – Simple Linear Regression: training scatter + fitted line.  
2) Figure 2 – Predicted vs Actual for the **best** model (by RMSE then MAE).

## 1. Imports & Setup

```python
# Keep imports minimal and aligned with class examples
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from pathlib import Path
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

RANDOM_STATE = 42  # reproducibility

OUTPUT_DIR = Path('/mnt/data')
FIG_DIR = OUTPUT_DIR / 'task1_figures'
TABLE_DIR = OUTPUT_DIR / 'task1_tables'
FIG_DIR.mkdir(parents=True, exist_ok=True)
TABLE_DIR.mkdir(parents=True, exist_ok=True)
```

## 2. Load Data

```python
# Expect the CSV to be in the same folder as this notebook
csv_path = Path('houseprice_data.csv')
assert csv_path.exists(), "houseprice_data.csv not found. Place it next to this notebook."
df = pd.read_csv(csv_path)

# Basic checks
assert 'price' in df.columns, "Target column 'price' is missing."
print('Shape:', df.shape)
df.head()
```

## 3. Feature Preparation (No correlation table exported)

```python
# Use numeric columns only; target is 'price'
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
feature_cols = [c for c in numeric_cols if c != 'price']

X_all = df[feature_cols].values
y = df['price'].values

# Internally determine the top features by absolute Pearson correlation with price
corr = df[feature_cols + ['price']].corr()['price'].drop('price')
corr_sorted = corr.reindex(corr.abs().sort_values(ascending=False).index)

# Pick top 1, top 2, and top 5
top1 = corr_sorted.index[0]
top2 = corr_sorted.index[1] if len(corr_sorted) > 1 else corr_sorted.index[0]
top5 = list(corr_sorted.index[:5]) if len(corr_sorted) >= 5 else list(corr_sorted.index[:len(corr_sorted)])

print('Selected single feature:', top1)
print('Selected two features:', [top1, top2])
print('Selected top-5 features:', top5)
```

## 4. Train/Test Split (consistent across models)

```python
# Design matrices for each model
X_A = df[[top1]].values               # Model A: 1 feature
X_B = df[[top1, top2]].values         # Model B: 2 features
X_C = df[top5].values                 # Model C: top 5 features
X_D = X_all                           # Model D: all numeric features

# Same split for fairness
X_train_A, X_test_A, y_train, y_test = train_test_split(X_A, y, test_size=0.2, random_state=RANDOM_STATE)
X_train_B, X_test_B, _, _ = train_test_split(X_B, y, test_size=0.2, random_state=RANDOM_STATE)
X_train_C, X_test_C, _, _ = train_test_split(X_C, y, test_size=0.2, random_state=RANDOM_STATE)
X_train_D, X_test_D, _, _ = train_test_split(X_D, y, test_size=0.2, random_state=RANDOM_STATE)
```

## 5. Train & Evaluate the Four Models

```python
def fit_eval(X_train, X_test, y_train, y_test):
    model = LinearRegression()
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    rmse = mean_squared_error(y_test, preds, squared=False)
    mae  = mean_absolute_error(y_test, preds)
    r2   = r2_score(y_test, preds)
    return model, preds, {'RMSE': rmse, 'MAE': mae, 'R2': r2}

lin_A, yA, metA = fit_eval(X_train_A, X_test_A, y_train, y_test)
lin_B, yB, metB = fit_eval(X_train_B, X_test_B, y_train, y_test)
lin_C, yC, metC = fit_eval(X_train_C, X_test_C, y_train, y_test)
lin_D, yD, metD = fit_eval(X_train_D, X_test_D, y_train, y_test)

# Build comparison table
results_df = pd.DataFrame([
    {'Model': f'Model A – Single-feature LR ({top1})', **metA},
    {'Model': f'Model B – Two-feature LR ({top1}, {top2})', **metB},
    {'Model': f'Model C – Top-5-features LR ({", ".join(top5)})', **metC},
    {'Model': 'Model D – All-features LR', **metD},
]).round({'RMSE':0,'MAE':0,'R2':3})

# Save comparison table for the report
table_path = TABLE_DIR / 'task1_model_comparison.csv'
results_df.to_csv(table_path, index=False)
print('Saved comparison table:', table_path)

results_df
```

### Figure 1 – Simple Linear Regression: training scatter + fitted line

```python
# Visualise the relationship for Model A (single feature) on the training split
plt.figure()
plt.scatter(X_train_A, y_train, s=10)
xs = np.linspace(X_A.min(), X_A.max(), 200).reshape(-1, 1)
ys = lin_A.predict(xs)
plt.plot(xs, ys)
plt.xlabel(top1)
plt.ylabel('price')
plt.title('Figure 1: Simple Linear Regression – Training Scatter + Fitted Line')
fig1_path = FIG_DIR / 'figure1_simple_linear_regression.png'
plt.savefig(fig1_path, bbox_inches='tight')
print('Saved:', fig1_path)
plt.show()
```

### Figure 2 – Predicted vs Actual (best model)

```python
# Choose best by RMSE, then MAE
ordered = results_df.sort_values(['RMSE','MAE'], ascending=[True, True]).reset_index(drop=True)
best_name = ordered.loc[0, 'Model']

# Select predictions for the best model
if best_name.startswith('Model A'):
    y_best = yA
elif best_name.startswith('Model B'):
    y_best = yB
elif best_name.startswith('Model C'):
    y_best = yC
else:
    y_best = yD

plt.figure()
plt.scatter(y_test, y_best, s=10)
mn, mx = float(min(y_test.min(), y_best.min())), float(max(y_test.max(), y_best.max()))
plt.plot([mn, mx], [mn, mx])  # ideal reference line
plt.xlabel('Actual price')
plt.ylabel('Predicted price')
plt.title('Figure 2: Predicted vs Actual – Best Model')
fig2_path = FIG_DIR / 'figure2_pred_vs_actual_best.png'
plt.savefig(fig2_path, bbox_inches='tight')
print('Saved:', fig2_path)
plt.show()

ordered
```
