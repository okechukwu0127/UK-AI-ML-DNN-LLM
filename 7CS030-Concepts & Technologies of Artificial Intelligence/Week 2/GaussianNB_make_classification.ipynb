#### This code demonstrates binary classification using the Naive Bayes (GaussianNB) model, applied to a synthetic dataset generated with make_classification.

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt # MATLAB-like way of plotting

# sklearn package for machine learning in python:
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.datasets import make_classification
```

- Generates 400 samples, with 1 feature and 2 classes.

- flip_y=0.05 introduces 5% label noise (adds some randomness).

 - random_state=4 ensures reproducibility.

```python
# generate regression data set
X, y = make_classification(
    n_samples=400, n_features=1,
    n_redundant=0, n_informative = 1, n_classes=2,
    n_clusters_per_class=1, flip_y=0.05, random_state=42
)
```

```python
# split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 1/3, random_state=0)
```

```python
gnb = GaussianNB()
gnb.fit(X_train, y_train)
```

```python
# accuracy score
print('Our Accuracy is %.2f' % gnb.score(X_test, y_test))

# number of mislabeled points
print('Number of mislabeled points out of a total %d points : %d'
% (X_test.shape[0], (y_test != gnb.predict(X_test)).sum()))
```

```python
# visualise the model
fig1, ax1 = plt.subplots()
ax1.scatter(X_test, y_test, color='blue')
ax1.scatter(X_test, gnb.predict(X_test), color='red', marker='*')
ax1.scatter(X_test, gnb.predict_proba(X_test)[:,1], color='green', marker='.')
ax1.set_xlabel('X')
ax1.set_ylabel('y')
```

### Conclusion

- GaussianNB works well for problems with normally distributed data.

- With 1 feature, it’s ideal for visual explanation and performance tracking.

- The fewer the mislabeled points, the better your model generalizes.
