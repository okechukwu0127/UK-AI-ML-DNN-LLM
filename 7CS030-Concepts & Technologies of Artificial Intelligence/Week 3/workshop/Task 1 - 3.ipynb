#### T1 Use the K-means algorithm to cluster the 8 data points given below into K = 3 clusters.

The coordinates of the data points are: 
 - x(1) = (2, 8), x(2) = (2, 5), x(3) = (1, 2), x(4) = (5, 8),\n
 - x(5) = (7, 3), x(6) = (6, 4), x(7) = (8, 4), x(8) = (4, 7),

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import Normalize

from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from sklearn.cluster import MeanShift 
from sklearn.preprocessing import LabelEncoder
```

```python
X = np.array([[2,8],[2,5],[1,2],[5,8],[7,3],[6,4],[8,4],[4,7]])
X
```

```python
kmeans = KMeans(n_clusters = 3, n_init = 'auto', random_state = 42)
kmeans.fit(X)
centers = kmeans.cluster_centers_
```

```python
fig, ax = plt.subplots()

nm = Normalize(vmin = 0, vmax = len(centers)-1)

ax.scatter(X[:,0], X[:,1], c = kmeans.predict(X), s = 50, cmap = 'plasma', norm = nm)

for i in range(centers.shape[0]):
    ax.text(centers[i,0], centers[i,1], str(i), c = 'black',
            bbox = dict(boxstyle = "round", facecolor = "white", edgecolor = "black"))

ax.set_xlabel('X1')
ax.set_ylabel('X2')
```

#### T2 Use the K-Means algorithm to cluster the data generated by the make_blobs func-tion in Python. Construct data sets with

#### a) 400 points, 2 clusters with standard deviation 0.7

```python
X, y = make_blobs(n_samples = 400, n_features = 2, centers = 2, cluster_std = 0.7, random_state = 42)
#print(X,y) #Can you determine what y represents here?

kmeans = KMeans(n_clusters = 2, n_init = 'auto', random_state = 42)
kmeans.fit(X)
centers = kmeans.cluster_centers_
print('Centroids :\n', centers)
```

#### b) 1000 points, 3 clusters with standard deviation 1.0

```python
X, y = make_blobs(n_samples = 100, n_features = 2, centers = 3, cluster_std = 1.0, random_state = 42)

kmeans = KMeans(n_clusters = 3, n_init = 'auto', random_state = 42)
kmeans.fit(X)
centers = kmeans.cluster_centers_
print('Centroids :\n', centers)
```
