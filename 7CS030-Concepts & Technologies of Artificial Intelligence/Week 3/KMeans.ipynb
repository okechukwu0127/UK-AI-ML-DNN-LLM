**K-Means clustering** is an unsupervised learning algorithm — it finds structure in unlabeled data by grouping similar points.

### Purpose of the Code
This code demonstrates how to:

 - Cluster a small set of 2D data points into groups using K-Means.
 - Identify the cluster centroids (the mean position of each group).
 - Predict which cluster a new test point belongs to.
 - Visualize the results using a scatter plot, coloring points by their cluster and labeling centroids.

```python
import numpy as np  # Import NumPy for numerical computations and handling arrays
import pandas as pd  # Import pandas (not used here, but often used for data handling)
import matplotlib.pyplot as plt  # Import Matplotlib for plotting and visualization
from matplotlib.colors import Normalize  # Used to normalize color mapping for plots
from sklearn.cluster import KMeans  # Import KMeans clustering algorithm from scikit-learn
```

```python
# Define a small 2D dataset (each row is a data point with [x, y] coordinates)
X = np.array([[2, 3], [6, 1], [1, 2], [3, 0]])
```

```python
# Display the dataset in the console
print('Data:\n', X, '\n')
```

```python
# Create a KMeans model that will try to form 2 clusters
# n_clusters = 2 → we want to split the data into 2 groups
# n_init='auto' → automatically decides how many times to run the algorithm with different initializations
# random_state=5 → ensures reproducibility (same results every run)
kmeans = KMeans(n_clusters=2, n_init='auto', random_state=5)
```

```python
# Fit (train) the KMeans model on the dataset X
# The algorithm will assign each point to one of the 2 clusters and find optimal centroid positions
kmeans.fit(X)
```

```python
# Retrieve the coordinates of the cluster centroids
centers = kmeans.cluster_centers_
print('Centroids:\n', centers, '\n')
```

```python
# Define a new data point for testing
test_point = np.array([[3, 1]])
```

```python
# Predict which cluster the new test point belongs to (0 or 1)
print('Prediction:', kmeans.predict(test_point), '\n')
```

```python
# Create a figure and axis for plotting
fig, ax = plt.subplots()

# Create a normalization object to ensure cluster colors are scaled consistently
nm = Normalize(vmin=0, vmax=len(centers)-1)

# Plot the test point (the one we predicted) using a star (*) marker
# It is colored according to its predicted cluster
ax.scatter(test_point[:, 0], test_point[:, 1], marker='*',
           c=kmeans.predict(test_point), s=50, cmap='plasma', norm=nm)

# Plot each centroid and label it with its cluster index
for i in range(centers.shape[0]):
    ax.text(centers[i, 0], centers[i, 1], str(i), c='black',
            bbox=dict(boxstyle="round", facecolor='white', edgecolor='black'))


# Label the axes
ax.set_xlabel('X1')
ax.set_ylabel('X2')
```

### Summary of What the Code Achieves

**Input:** A small dataset of 4 points in 2D space.

**Process:**
 - The K-Means algorithm divides these points into 2 clusters.
 - Each cluster has a centroid (mean of points in that group).
 - The algorithm assigns each point to its nearest centroid.

**Output:**
 - The coordinates of the 2 centroids.
 - The predicted cluster for a new point [3,1].
 - A scatter plot saved as cluster_plot.png showing:
    - Original data points colored by cluster.
    - Centroids labeled 0 and 1.
    - The test point shown as a star.
