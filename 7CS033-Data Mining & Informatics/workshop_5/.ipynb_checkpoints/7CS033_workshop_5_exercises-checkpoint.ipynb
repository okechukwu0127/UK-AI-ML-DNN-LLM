## Importing necessary Libraries  

```python
import pandas as pd
import numpy as np
```

##### Q1: Write code to load the transaction_100.csv dataset and display the first 5 rows. Determine how many transaction there are and extract a list of the unique items in the dataset.

```python
#Solution

df = pd.read_csv('transaction_100.csv')

print(df.head())
print('Number of transactions: ', df.shape[0])
```

```python
#We can see that the items are currently in lists seperated by semicolons. Lets plit those and then get a list of the unique items

df['Item_List'] = df['Items'].apply(lambda x: x.split(';'))
print(df)
#get the unique items 
print(np.unique(df['Item_List'].sum()))
```

##### Q2: Write code to convert the transaction data into one-hot encoded format. You may find the TransactionEncoder function from mlxtend.preprocessing useful.

```python
#solution
from mlxtend.preprocessing import TransactionEncoder

transactions = df['Item_List'].tolist()

te = TransactionEncoder()
encoded_transactions = te.fit(transactions).transform(transactions)

encoded_df = pd.DataFrame(encoded_transactions, columns = te.columns_)

encoded_df.head()
```

##### Q3: Write code to apply the Apriori algorithm with minimum support = 0.13. Display all frequent itemsets and sort the results in descending order.

```python
#solution
from mlxtend.frequent_patterns import apriori

frequent_itemsets = apriori(encoded_df, min_support = 0.13, use_colnames =True)

print(frequent_itemsets.sort_values(by = 'support', ascending = False)) 
```

##### Q4: Write code to generate association rules using a minimum confidence = 0.7. Display the rules and sort by lift.

```python
#solution

from mlxtend.frequent_patterns import association_rules

rules = association_rules(frequent_itemsets, metric = 'confidence', min_threshold = 0.7)

print(rules.sort_values(by = 'lift', ascending = False))
```

##### Q5: Chose one rule with high lift and interpret that rule.

##### Solution:

The rule with the highest confidence and lift values is (Milk, Beer) => Diapers. This rule states that customers who buy milk and beer and highly likley to purchase diapers. The lift value is 3.33, as the lift is much greater than 1, this tells us that there is strong evidence that this combination occurs more frequently than we would have expected if the santecendents and consequents were independant.

##### Q6: Write code to re-run the Apriori algorithm with a variety of minimum support values. Compare the number of frequent itemsets, what happens as the threshold is increased?

```python
#Solution

#There are a number of ways this could be done. However, it is useful to use a for loop to try a variety of minimum support values quickly.

support_values = np.arange(0.1, 0.6, 0.05)

results = []

for i in support_values:
    frequent_itemsets = apriori(encoded_df, min_support = i, use_colnames = True)

    results.append({'min_support': i, 'num_itemsets': frequent_itemsets.shape[0]})

results_df = pd.DataFrame(results)
results_df
```

```python
#We could also visualise this

import matplotlib.pyplot as plt

plt.plot(results_df['min_support'], results_df['num_itemsets'], marker = 'o')
plt.xlabel('Minimum Support')
plt.ylabel('Number of Frequent Itemsets')
plt.title('Effect of Support Threshold on Itemsets')
plt.show()
```

As minimum support increases the number of frequent itemsets decreases as more itemsets are pruned at higher thresholds.

##### Q7: Why could a high confidence value alone not necessarily imply a strong association rule?

##### Solution:

Lets motivate this with an example. Suppose the rule Cereal => Milk is found to have confidence = 0.85 and lift = 1.02.

If milk is a common itemset in the data, for example 75% of all transactions contain milk. Then even randomly selected transactions are likley to contain milk. This does not necessarily mean that buying cereal casues people to buy milk. 

Suppose:
- Support(Cereal) = 0.2
- Support(Milk) = 0.75
- Support(Cereal, Milk) = 0.17

Then confidence = 0.17/0.2 = 0.85. Which implies a strong relationship. But lift = 0.85/0.17 = 1.13, a lift value close to 1 means the probability of buying milk given you have brought cereal is approximatley the same as the probability of just buying milk. The value of 1.13 implies buying cereal does not increase the probability of buying milk by much. This is why in practice it is useful to look beyond confidence to determine if the association rules are appropriate.
