```python
############# WRITE YOUR CODE IN THIS CELL (IF APPLICABLE)####################
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
#import portugal, sweden and uk dataset
df_portugal = pd.read_csv('Portugal_online_retail.csv', dtype={'InvoiceNo': str})
df_sweden = pd.read_csv('Sweden_online_retail.csv', dtype={'InvoiceNo': str})
df_uk = pd.read_csv('UK_online_retail.csv', dtype={'InvoiceNo': str})
```

```python
#Drop the InvoiceNo column
df_portugal = df_portugal.drop('InvoiceNo', axis =1);
df_sweden = df_sweden.drop('InvoiceNo', axis =1);
df_uk = df_uk.drop('InvoiceNo', axis =1);
```

```python
# Convert 0/1 to boolean (VERY IMPORTANT)
df_portugal = df_portugal.astype(bool)
df_sweden = df_sweden.astype(bool)
df_uk = df_uk.astype(bool)
```

```python
print(f"{df_portugal.shape[0]} = Portugal rows!")
print(f"{df_sweden.shape[0]} = Sweden rows!")
print(f"{df_uk.shape[0]} = UK rows!")
```

```python
#Since Portugal and Sweden are very small datasets, we must use lower support/confidence.
#For UK (large dataset), we can use higher thresholds.
```

### PORTUGAL - RUN APRIORI (Different support per dataset)

```python
#For Small datase, support cannot be too high (you will lose patterns)
#For Large dataset, support must be lower to capture frequent combinations
```

##### GENERATE FREQUENCY WITH 3 DIFFERENT SUPPORT LEVELS

```python
# Test three support values
frequent_p_01 = apriori(df_portugal, min_support=0.04, use_colnames=True);
frequent_p_05 = apriori(df_portugal, min_support=0.05, use_colnames=True);
frequent_p_10 = apriori(df_portugal, min_support=0.10, use_colnames=True);
```

```python
print("Number of Frequent Itemsets for 0.04:", len(frequent_p_01))
print("Number of Frequent Itemsets for 0.05:", len(frequent_p_05))
print("Number of Frequent Itemsets for 0.10:", len(frequent_p_10))
```

```python
#For the Portugal dataset, a support of 0.10 is best because it produces a manageable number of 
#meaningful itemsets, while lower support values generate too many noisy itemsets
```

##### GENERATE RULES WITH 3 DIFFERENT CONFIDENCE LEVELS

```python
# Generate association rules (try different confidence levels)
rules_03 = association_rules(frequent_p_10, metric="confidence", min_threshold=0.3)
rules_05 = association_rules(frequent_p_10, metric="confidence", min_threshold=0.5)
rules_07 = association_rules(frequent_p_10, metric="confidence", min_threshold=0.7)

print("Confidence 0.3:", len(rules_03))
print("Confidence 0.5:", len(rules_05))
print("Confidence 0.7:", len(rules_07))
```

```python
#The best association rules for Portugal are those with a confidence threshold of 0.5 because they balance 
#reliability and quantity, capturing meaningful patterns without being too noisy or too restrictive.
```

```python
# Select confidence = 0.5
rules_portugal = rules_05

# Sort by lift (higher lift = stronger rule)
top_portugal = rules_portugal.sort_values(by='lift', ascending=False).head(3)
top_portugal
```

### SWEDEN - RUN APRIORI (Different support per dataset)

##### GENERATE FREQUENCY WITH 3 DIFFERENT SUPPORT LEVELS

```python
# Test three support values

frequent_s_08 = apriori(df_sweden, min_support=0.08, use_colnames=True)
frequent_s_10 = apriori(df_sweden, min_support=0.10, use_colnames=True)
frequent_s_15 = apriori(df_sweden, min_support=0.15, use_colnames=True)

print("Support 0.08:", len(frequent_s_08))
print("Support 0.10:", len(frequent_s_10))
print("Support 0.15:", len(frequent_s_15))
```

```python
#0.10 provided a balanced and meaningful number of itemsets (25) while 0.08 produced too many 
#frequent itemsets (118), indicating noise, while 0.15 produced too few (3)
```

##### GENERATE RULES WITH 3 DIFFERENT CONFIDENCE LEVELS

```python
# Generate association rules (try different confidence levels)
rules_s_04 = association_rules(frequent_s_10, metric="confidence", min_threshold=0.4)
rules_s_05 = association_rules(frequent_s_10,  metric="confidence", min_threshold=0.5)
rules_s_07 = association_rules(frequent_s_10, metric="confidence", min_threshold=0.7)


print("Confidence 0.4:", len(rules_s_04))
print("Confidence 0.5:", len(rules_s_05))
print("Confidence 0.7:", len(rules_s_07))
```

```python
#0.7 Confidence was selected because all six rules remained the same across 0.4, 0.5, and 0.7, meaning the 
#rules are already strong, so using the higher threshold ensures greater reliability without losing any rules
```

```python
# Select confidence = 0.5
rules_sweden = rules_s_05

# Sort by lift (higher lift = stronger rule)
top_sweden= rules_sweden.sort_values(by='lift', ascending=False).head(3)
top_sweden
```

### UK - RUN APRIORI (Different support per dataset)

##### GENERATE FREQUENCY WITH 3 DIFFERENT SUPPORT LEVELS

```python
### UK - TEST SUPPORT VALUES
frequent_uk_009 = apriori(df_uk, min_support=0.009, use_colnames=True)
frequent_uk_01 = apriori(df_uk, min_support=0.01, use_colnames=True)
frequent_uk_02 = apriori(df_uk, min_support=0.02, use_colnames=True)

print("Support 0.009:", len(frequent_uk_009))
print("Support 0.01:", len(frequent_uk_01))
print("Support 0.02:", len(frequent_uk_02))
```

```python
#Support = 0.01 is the best choice because it captures meaningful and recurring purchasing patterns without being 
#too restrictive or generating excessive noise. Lower support includes too many insignificant combinations, 
#while higher support removes useful associations
```

##### GENERATE RULES WITH 3 DIFFERENT CONFIDENCE LEVELS

```python
rules_uk_04 = association_rules(frequent_uk_01, metric="confidence", min_threshold=0.4)
rules_uk_05 = association_rules(frequent_uk_01, metric="confidence", min_threshold=0.5)
rules_uk_06 = association_rules(frequent_uk_01, metric="confidence", min_threshold=0.6)

print("Confidence 0.4:", len(rules_uk_04))
print("Confidence 0.5:", len(rules_uk_05))
print("Confidence 0.6:", len(rules_uk_06))
```

```python
# Select confidence = 0.5
rules_uk = rules_uk_05

# Sort by lift (higher lift = stronger rule)
top_uk= rules_uk.sort_values(by='lift', ascending=False).head(3)
top_uk
```
