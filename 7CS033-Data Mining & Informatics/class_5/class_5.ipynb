```python
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

```python

df = pd.read_csv('France_online_retail.csv')
df
```

```python
df = df.drop('InvoiceNo', axis =1);
df
```

```python
freq_items = apriori(df, min_support=0.12, use_colnames = True)
# the lesser the min_support, the more the values it displays 
#0.15
```

```python
#one hot encoding help you to make
```

```python
print(freq_items.shape)
```

```python
freq_items.sort_values(by=['support'], ascending=False)
```

```python
#rules = association_rules(freq_items,metric ='confidence', min_thresholds=0.6);
rules = association_rules(freq_items, metric='confidence', min_threshold=0.6)
```

```python
print(rules.shape)
```

```python
rules.sort_values(by=['confidence'], ascending=False)
```
