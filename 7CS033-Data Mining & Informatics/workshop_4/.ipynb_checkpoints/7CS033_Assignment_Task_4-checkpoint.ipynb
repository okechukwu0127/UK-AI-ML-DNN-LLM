## TASK 4.1:  Download the adult_T4 dataset. Apply K-Means and Hierarchical clustering to three optional columns in the dataset. Find the optimum number of clusters for both clustering methods (10%).

## NOTE: You should comment on your code wherever necessary and briefly explain what the code is doing

```python
############# WRITE YOUR CODE IN THIS CELL (IF APPLICABLE)####################

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv('adult_T4.csv')

d1 = df['age'] #target variable 1
d2 = df['education-num'] #target variable 2
d3 = df['hours-per-week'] #target variable 3
```

```python
#columns = ['age', 'hours-per-week', 'education-num']
#X = df[columns]

X = pd.concat([d1,d2,d3], axis = 1); #get our 3 input variables
X = X.dropna()
```

```python
#print(df_.head()) # shape of the dataset
#print(df_.shape)
#print(df_.info()) # basic information about each column
```

```python
#identify the number of missing values i
#print(pd.isnull(df_).sum());
```

```python

scaler = StandardScaler()
X = scaler.fit_transform(X) #scale the features of X

print(X)

#Scaling is important for clustering as distance is used to determine the clusters. If the features are not transformed 
#to be on the same scale then patterns in the data could be hidden by certain features dominating due to their different
#scale. Standardising all of the data ensures that each feature will fairly contribute to the clustering process.
```

```python
#Apply K-Means 

#Since we will need to plot the elbow curve we will create an empty list to store the inertia values
inertia = []

#Write code to fit K-means clustering models for values of k = 1 to k = 8 i.e 7
#Fit our model for the appropriate range of k and store the inertia values for later use
for k in range(1, 8):
    km = KMeans(n_clusters = k, random_state = 42)
    km.fit(X)
    inertia.append(km.inertia_)

#plot the elbow curve
plt.figure(figsize=(8,5))
plt.plot(range(1,8), inertia, marker = 'o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.show()
```

```python
#Based off of the plot we could argue that k = 2, k = 3 or k = 4 are appropriate, let us check with silhouette scores.

sil_scores = []

#
for k in range (2, 9):
    km = KMeans(n_clusters = k, random_state = 42)
    labels = km.fit_predict(X)
    score = silhouette_score(X, labels)
    sil_scores.append(silhouette_score(X, labels))
    print(f"k={k}, Silhouette Score={score}")


plt.figure(figsize = (8, 5))
plt.plot(range(2,9), sil_scores, marker = 'o')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Silhouette Score')
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans_predict =kmeans.fit_predict(X)
```

```python
optimal_k_kmeans = np.argmax(sil_scores) + 2
print("Optimal k for KMeans:", optimal_k_kmeans)
```

```python

#kmeans_predict
```

```python

#Hierarchical clustering 
linked = linkage(X, method='ward')

plt.figure(figsize=(10, 5))
dendrogram(linked)
plt.show()
```

```python
hierarchy = []

for k in range(2, 9):
    agg = AgglomerativeClustering(n_clusters=k, linkage='ward')
    labels = agg.fit_predict(X)
    hierarchy_score=silhouette_score(X, labels)
    hierarchy.append(hierarchy_score)
    print(f"k={k}, Silhouette Score={hierarchy_score}")


plt.figure(figsize=(8, 5))
plt.plot(range(2, 9), hierarchy, marker='o')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score for Agglomerative Clustering')
plt.show()
```

```python
optimal_k_hier = np.argmax(hierarchy) + 2
print("Optimal k for Hierarchical:", optimal_k_hier)
```

## Task 4.2: Apply the PCA method to the dataset and extract the first two principal components (n_components=2). Plot the scatter plot of the dataset's first two components for the two classes of the income column (5%). 

## NOTE 1: You should comment on your code wherever necessary and briefly explain what the code is doing.

```python


# Agglomerative clustering
agg = AgglomerativeClustering(n_clusters=3)
agg_predict = agg.fit_predict(X)

# PCA for visualization / dimensionality reduction
pca = PCA(n_components=2)
x_pca = pca.fit_transform(X)

#x_pca
```

```python

plt.figure(figsize=(8,5))
plt.scatter(x_pca[:,0], x_pca[:,1], c=kmeans_predict)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(x_pca[:,0], x_pca[:,1], c=agg_predict)
plt.show()
```
