### TASK 1.1: Write a code to receive an optional text, capitalise all words in the text and print them (2%)

```python
############# WRITE THE CODE IN THIS CELL ####################
#Solution

# Ask the user to enter text (optional input)
# If the user just presses Enter, use a default sentence
text1 = input("Enter a sentence: ")

# Create an empty list to store capitalized words
output=[];

if text1 == "":
    text1 = "welcome to assessment one"

# Split the sentence into individual words
for word in text1.split():

    # Capitalize the first letter of each word
    capitalized_word= word.capitalize()

    # Add the capitalized word to the list
    output.append(capitalized_word)
    #print(word);

print(output)

# Join the words back into a full sentence
sentence = ' '.join(output);

# Print the final result
print("Capitalized Sentence:", sentence)

```

### TASK 1.2: Write a program that prompts the user to input the radius of a circle and calculates its area (2%)

```python
############# WRITE THE CODE IN THIS CELL ####################

# Import the math module to use the value of pi
import math


# Prompt the user to enter the radius (allow decimal values)
inputRadius = float(input("Enter Radius of the circle: "))

# Calculate the area using the formula: πr²
circleArea = math.pi * (inputRadius ** 2)


# Display the result
print("The area of the circle is:", circleArea)

```

### TASK 1.3: Download and import the 'adult' dataset from Canvas. Write a function to split the dataset column-wise into two halves and swap the first half with the second half (3%)

```python
############# WRITE THE CODE IN THIS CELL ####################

import pandas as pd
import numpy as np

# Read the CSV file into a DataFrame
df = pd.read_csv('adult.csv');


# Define a function to split and swap columns
def swap_column_halves(df):

    # Get total number of columns
    num_cols = df.shape[1]
    
    # Find the midpoint of the columns
    mid = num_cols // 2
    
    # Split the columns into two halves
    first_half = df.iloc[:, :mid]
    second_half = df.iloc[:, mid:]
    
    # Swap the halves and concatenate them
    swapped_df = pd.concat([second_half, first_half], axis=1)

    return swapped_df



# Call the function
swapped_df = swap_column_halves(df)

# Display the swapped dataframe
swapped_df



```

### Task 1.4: Write a function that takes the names of two numerical columns as input and compares their values for all rows. If the value in the first column is greater than the value in the second column, the function should return True; otherwise, it should return False. The function should append a new column to the dataset to store the comparison results for all rows. Apply the function to the 'age' and 'hours-per-week' columns in the adult dataset and print the resulting dataset (3%)

```python
############# WRITE THE CODE IN THIS CELL ####################

df = pd.read_csv('adult.csv');



# Define a function that:
# 1. Takes a dataframe
# 2. Takes two numerical column names as input
# 3. Compares their values row-by-row
# 4. Appends a new column with True/False results
def isGreaterThan(dataframe, col1, col2):
    
    # Perform vectorized comparison:
    # Returns True if value in col1 > value in col2, otherwise False
    dataframe['isGreaterThan'] = dataframe[col1] > dataframe[col2]

    # Return the updated dataframe
    return dataframe

# Apply the function to compare 'age' and 'hours-per-week'
df = isGreaterThan(df, 'age', 'hours-per-week')

# Display the resulting dataset
df
```
