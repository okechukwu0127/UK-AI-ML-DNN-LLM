```python
import numpy as np #maths package
import pandas as pd #used to manipulate our data


```

```python
#data frame
#since the data does not have a header we will tell pandas that ther is no header
iris_df = pd.read_csv('iris.data',header=None);
```

```python
print(iris_df.head()); #Tells us the colums data
print(iris_df.dtypes); # tells us data type for each colums
print(iris_df.shape)
```

```python
iris_df.columns =["sapal_length","sapal_width", "petal_length","petal_width","species"];
print(iris_df.head());
```

```python
#Rename Columns
iris_df = iris_df.rename(columns ={
"sapal_length":"sapal_length_cmm","sapal_width":"sapal_width_cmm"
    
})
```

```python
print(iris_df.head());
```

```python
#change columns location
iris_df = iris_df.iloc[:,[4,0,1,2,3]]

print(iris_df.head());
```

```python
#add a new column
iris_df['petal_area_cm2']= iris_df['petal_length'] *  iris_df['petal_width'];
print(iris_df.head());
```

```python
#Demo a for loop
size = []


for i in range(len(iris_df)):
    #categorize petals based on widths and lengh
    petal_length_cm = iris_df.iloc[i][3] # were 3 is the 4th column
    petal_width_cm = iris_df.iloc[i][4]


    if petal_length_cm > 4 and petal_width_cm > 1.5:
        size.append('Lagrge');
    else:
        size.append('Small');

iris_df['size'] = size;
```
