```python
#we need our taget ana our input data
X = df.drop('Diagnosis',axis=1) # our  input data
y = df.iloc[:, 30] #locate colum 30 to get out target data
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()   # create scaler object

x = scaler.fit_transform(X)

x
```

```python


from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y = le.fit_transform(y)
print(y)
```

```python
from sklearn.model_selection import train_test_split
TEST_SIZE = 0.15
SEED = 42;
#3

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=TEST_SIZE, random_state=SEED
)
```

```python
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier 
from sklearn.neighbors import KNeighborsClassifier 
from sklearn.tree import DecisionTreeClassifier

#lets call all our models we want to use
SVM =svm.SVC()
RF = RandomForestClassifier()
KNN = KNeighborsClassifier() 
DT = DecisionTreeClassifier()
```

```python
#Lets 
SVM.fit(X_train, y_train)
RF.fit(X_train, y_train)
KNN.fit(X_train, y_train)
DT.fit(X_train, y_train)
```

```python
#we need to evalaute our models, we do that by:
#The prediction each model make on our test data, we would use it to create our confusion metrix
y_pred1 = SVM.predict(X_test) 
y_pred2 = RF.predict(X_test) 
y_pred3 = KNN.predict(X_test) 
y_pred4 = DT.predict(X_test)
```

```python

import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm1 = confusion_matrix(y_test, y_pred1, labels = SVM.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix = cm1, display_labels = SVM.classes_) 
disp.plot()
plt.title('SVM')
```

```python
#the true label is zero (yellow - top left)
#the predicted label is zero (purple - bottom left)
```

```python
#if the true label is 0 and our prediction label is 0 , this are true positives
#if the true label is 0 and our prediction label is 1 then its a false negative
```

```python
cm2 = confusion_matrix(y_test, y_pred2, labels = SVM.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix = cm1, display_labels = SVM.classes_) 
disp.plot()
plt.title('SVM')
```

```python
from sklearn import metrics

def confusion_metrics (y_pred):
    conf_accuracy = metrics.accuracy_score(y_test, y_pred)
    conf_recall = metrics.recall_score(y_test, y_pred)
    conf_specificity = metrics.recall_score(y_test, y_pred, pos_label= 0)
    conf_precision = metrics.precision_score(y_test, y_pred)
    conf_f1 = metrics.f1_score(y_test, y_pred)
    print('-'*50)
    print('Accuracy: %.2f' %conf_accuracy)
    print('Recall: %.2f' %conf_recall)
    print('Specificity: %.2f' %conf_specificity)
    print ('Precision: %.2f' %conf_precision)
    print('F1: %.2f' %conf_f1)
    print('\n \n')
```

```python
confusion_metrics(y_pred1)
confusion_metrics(y_pred2)
confusion_metrics(y_pred3)
confusion_metrics(y_pred4)
```
