##### Q5: Load the bankdata.csv dataset, change the display settings so that the maximum number of columns displayed when printing a DataFrame is 17. 

```python
#Solution

bankdf = pd.read_csv('bankdata.csv')

pd.set_option('display.max_columns', 17)

bankdf
```

##### Q6: Write code to explore the data and identify the data types and any columns with missing values.

```python
#Solution

print(bankdf.info())
print(bankdf.isna().sum())
print(bankdf.sample(20))
# We can see that there are a number of missing values for age. Some features do not having missing values, but do have 
# 'unknown' entries.
```

##### Q7: Write code to replace the missing values in the 'age' column with the mean age.

```python
#Solution
mean_age = bankdf['age'].mean(skipna = True)

print(mean_age)

bankdf['age'].fillna(mean_age, inplace = True)

print(bankdf.isna().sum())
```

##### Q8: Write code to replace the 'unknown' values in the data with NaN.

```python
#Solution

bankdf.replace('unknown', pd.NA, inplace = True)

print(bankdf.isna().sum())
```

##### Q9: Write code to remove the 'poutcome' and 'contact' columns. Appropriately handle the missing values in the education column, justify your choice.

```python
#Solution

bankdf = bankdf.drop(['contact', 'poutcome'], axis = 1)

print(bankdf.isna().sum())

bankdf['education'] = bankdf['education'].fillna(bankdf['education'].mode()[0])
bankdf['job'] = bankdf['job'].fillna(bankdf['job'].mode()[0])

print(bankdf.isna().sum())

# The mode was used for the categorical variables to preserve the most frequent category without reducing the dataset size.
```

##### Q10: Using the LabelEncoder method from sklearn, write code to encode the target feature 'y'.

```python
#Solution

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
bankdf['y'] = le.fit_transform(bankdf['y'])

bankdf.head(20)
```

##### Q11: Using the MinMaxScaler method from sklearn, write code to normalise the 'duration' feature so that the values like in the range [0,1].

```python
#Solution

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range = (0, 1))
bankdf['duration'] = scaler.fit_transform(bankdf[['duration']])

bankdf.head(10)
```
