```python
############# WRITE YOUR CODE IN THIS CELL (IF APPLICABLE)####################

import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
```

```python
#Change display setting and print first 100 rows

# Set pandas to display all columns
pd.set_option('display.max_columns', None)

# Set pandas to display up to 100 rows
pd.set_option('display.max_rows', 100)

# Load the dataset (ensure adult.csv is in your directory)
adult_df = pd.read_csv('adult_T3.csv')

# Print first 100 rows
print("First 100 rows of the dataset:")
print(adult_df.head(100))
```

```python
#Print dataset information
print("Dataset Information:")
print(adult_df.info())
```

### Handle the Null values

```python
# Replace '?' with NaN so pandas recognises them as missing
adult_df.replace('?', np.nan, inplace=True)

# Display null values BEFORE cleaning
print("Null values per column BEFORE cleaning:")
print(adult_df.isna().sum())
```

### Fill Missing Values

```python
#Select numerical columns
num_cols = adult_df.select_dtypes(include=['int64', 'float64']).columns

#Fill missing numerical values with mean
for col in num_cols:
    adult_df[col] = adult_df[col].fillna(adult_df[col].mean())
```

```python
# Select Text  columns
cat_cols = adult_df.select_dtypes(include=['object']).columns

# Fill missing Text values with mode (most frequent value)
for col in cat_cols:
    adult_df[col] = adult_df[col].fillna(adult_df[col].mode()[0])
```

```python
#Numerical missing values were replaced with the mean, and text-based missing values were 
#replaced with the mode to preserve data distribution and avoid losing records
```

```python
# Verify that no missing values remain
print("Null values per column AFTER cleaning:")
print(adult_df.isna().sum())
```

### Encode the Data

```python
#Encode the data
# Create LabelEncoder object
le = LabelEncoder()

# Encode target variable 'income'
adult_df['income'] = le.fit_transform(adult_df['income'])
```

```python
# Encode remaining Text columns
for col in cat_cols:
    if col != 'income':   # income already encoded
        adult_df[col] = le.fit_transform(adult_df[col])

print("Dataset after encoding:")
print(adult_df.head())
```

### Normalise the Data

```python
# Create scaler object
scaler = MinMaxScaler(feature_range=(0, 1))

# Apply scaling
scaled_data = scaler.fit_transform(adult_df)
```

```python
# Convert back to DataFrame
adult_scaled_df = pd.DataFrame(scaled_data, columns=adult_df.columns)

print("First 5 rows after normalisation:")
print(adult_scaled_df.head())
```
