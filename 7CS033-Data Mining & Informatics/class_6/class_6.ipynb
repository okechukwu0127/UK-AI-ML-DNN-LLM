```python
import pandas as pd
import numpy as np
```

```python
text = ["This is the NLP lecture by Jordan **",
        "IN this lectures I'll be explaining various DATA-CLEANING Techniques",
        "I hope you are PAYing attention! &&&",
        "Creating example text is pretty boring",
        "wlv ftw",
        "Running bulls cause chaotic situations"]

df = pd.DataFrame({'text': text})
df
```

```python
df = pd.DataFrame({'text':text})
df
```

```python
df['lower'] = df['text'].apply(lambda x: " ".join(x.lower() for x in x.split()))
```

```python
df
```

```python
import string
string.punctuation
```

```python
'!"#$%^%^&*(+_)(*&^%#@!
```

```python
df['punc1'] = df['lower'].apply(lambda x: " ".join(i for i in x if i not in string.punctuation))
```

```python
df
```

```python
import re
```

```python
df['punc2'] = df['lower'].apply(lambda x: "".join(re.findall(r'[a-zA-Z+" "]',x))) #remove other characters
```
